```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Lendo arquivo csv
# Lendo o data frame e retornando apenas o nome do arco e total de episodios
one_piece_arcs_episodes = pd.read_csv(r"C:\Users\hiros\OneDrive\Documentos\Paradigmas\aula1710\OnePieceArcs.csv").filter(items=['Arc', 'TotalEpisodes', 'TotalChapters']) # usecols = ['Arc', 'TotalEpisodes'] usecols = Seleciona colunas espec√≠ficas
one_piece_arcs_episodes['Arc'] = one_piece_arcs_episodes['Arc'].map(lambda x: x.rstrip('Arc'))  
print("Arcos e total de episodios")
print(one_piece_arcs_episodes)
```

```python
# Lendo o data frame filtrado e retornando apenas o nome dos 3 maiores arcos
top_episodes_one_piece = one_piece_arcs_episodes.sort_values('TotalEpisodes', ascending=False).head(3).loc[:, one_piece_arcs_episodes.columns != 'TotalChapters']
print("3 maiores arcos do anime")
print(top_episodes_one_piece)

top_chapters_one_piece = one_piece_arcs_episodes.sort_values('TotalChapters', ascending=False).head(3).loc[:, one_piece_arcs_episodes.columns != 'TotalEpisodes']
print("3 maiores arcos do manga")
print(top_chapters_one_piece)
```

```python
# Criando um gr√°fico e mostrando quantos episodios tem cada arco
print("3 maiores arcos do anime e total de episodios")
top_episodes_one_piece.plot(kind = 'bar', x='Arc', rot=0, figsize=(10,5))
```

```python
print("3 maiores arcos do manga e total de capitulos")
top_chapters_one_piece.plot(kind = 'bar', x='Arc', rot=0, figsize=(10,5))
```

```python
print("Todos os arcos do anime")
one_piece_arcs_episodes.plot(kind = 'bar', x='Arc', figsize=(20,5))
```
